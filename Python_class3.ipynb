#### **str.split()**
To split a string in Python, you primarily use the built-in .split() method. This method breaks a string into a list of substrings based on a specified delimiter.

- str.split(separator=None, maxsplit=-1)
  - separator (optional): The delimiter to use for splitting.
If not specified (or None), any sequence of whitespace (spaces, tabs, newlines) is used as the separator, and empty strings resulting from the split are discarded.

  - If specified, the string is split at each occurrence of that exact separator string.

  - maxsplit (optional): The maximum number of splits to perform.
  - maxsplit=-1 (default) means all occurrences of the separator are split.
  - If a number n is specified, the string will be split at most n times, resulting in a list with a maximum of n+1 elements.

```python
text = "Python makes coding simple"
words = text.split()
print(words)
```

### **any() function**
- The any() function in Python returns True if at least one element of the iterable is true. If the iterable is empty, it returns False. This function is useful for checking if any element in a list, tuple, or other iterable meets a certain condition.


```python
numbers = [2, 5, 8, 12, 3]
result = any(num > 13 for num in numbers)
print(result)
```

#### 2 .Tuple (tuple):
 - An ordered collection of items, which can be of different types. Tuples are immutable.


```python
cordinates=(10,20)
print(type(cordinates))
print(cordinates)
```

```python
# Demonstrating Tuple Immutability
my_tuple = (1, 2, 3, 4)
print(f"Original tuple: {my_tuple}")

# Attempting to modify an element in the tuple (will raise an error)
try:
    my_tuple[0] = 10
except TypeError as e:
    print(f"Error when trying to modify a tuple element: {e}")

# Attempting to append an element to the tuple (will raise an error)
try:
    my_tuple.append(5)
except AttributeError as e:
    print(f"Error when trying to append to a tuple: {e}")

# Attempting to remove an element from the tuple (will raise an error)
try:
    my_tuple.remove(2)
except AttributeError as e:
    print(f"Error when trying to remove from a tuple: {e}")

# Reassigning the tuple variable to a new tuple (this is allowed, but it's a new tuple, not a modification of the old one)
my_tuple = (10, 20, 30)
print(f"Tuple after reassignment: {my_tuple}")
```

# **Operators**


### Operators in Python
- Python provides a variety of operators to perform different operations. Here are the most commonly used operators:


- 1. **Arithmetic Operators**
    - Addition (+): Adds two operands.
    - Subtraction (-): Subtracts the second operand from the first.
    - Multiplication (*): Multiplies two operands.
    - Division (/): Divides the first operand by the second.
    - Modulus (%): Returns the remainder of the division.
    - Exponentiation (**): Raises the first operand to the power of the second.
    - Floor Division (//): Divides the first operand by the second and returns the largest integer less than or equal to the result.

```python
print(2**3) #2*2*2
```

```python
print(9/2)
```

```python
print(9//2)
```

```python
a = 10
b = 3

addition = a + b
subtraction = a - b
multiplication = a * b
division = a / b
modulus = a % b
exponentiation = a ** b
floor_division = a // b
print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Division:", division)
print("Modulus:", modulus)
print("Exponentiation:", exponentiation)
print("Floor Division:", floor_division)
```

- 2. **Comparison Operators**
  - Equal (==): Checks if two operands are equal.
  - Not Equal (!=): Checks if two operands are not equal.
  - Greater Than (>): Checks if the left operand is greater than the right.
  - Less Than (<): Checks if the left operand is less than the right.
  - Greater Than or Equal To (>=): Checks if the left operand is greater than or equal to the right.
  - Less Than or Equal To (<=): Checks if the left operand is less than or equal to the right.


```python
a = 10
b = 3

equal = (a == b)
not_equal = (a != b)
greater_than = (a > b)
less_than = (a < b)
greater_than_or_equal = (a >= b)
less_than_or_equal = (a <= b)
print("Equal:", equal)
print("Not Equal:", not_equal)
print("Greater Than:", greater_than)
print("Less Than:", less_than)
print("Greater Than or Equal:", greater_than_or_equal)
print("Less Than or Equal:", less_than_or_equal)
```

- 3.**Logical Operators**
    - AND (and): Returns True if both operands are true.
    - OR (or): Returns True if at least one operand is true.
    - NOT (not): Reverses the logical state of its operand.

```python
a = True
b = False

logical_and = a and b
logical_or = a or b
logical_not = not a
x=a and (not b)
print("x:",x)
print("Logical AND:", logical_and)
print("Logical OR:", logical_or)
print("Logical NOT:", logical_not)
```

- 4. **Assignment Operators**

  - Add and Assign (+=): Adds and assigns the result.
  - Subtract and Assign (-=): Subtracts and assigns the result.
  - Multiply and Assign (*=): Multiplies and assigns the result.
  - Divide and Assign (/=): Divides and assigns the result.
  - Modulus and Assign (%=): Takes modulus and assigns the result.
  - Exponent and Assign (**=): Raises to the power and assigns the result.
  - Floor Divide and Assign (//=): Floor divides and assigns the result.

```python
a = 10
print("Updated value of a:", a)
a += 5  # a = a + 5
print("Updated value of a:", a)
a -= 3  # a = a - 3
print("Updated value of a:", a)
a *= 2  # a = a * 2
print("Updated value of a:", a)
a /= 4  # a = a / 4
print("Updated value of a:", a)
a %= 3  # a = a % 3
print("Updated value of a:", a)
a **= 2  # a = a ** 2
print("Updated value of a:", a)
a //= 2  # a = a // 2
print("Updated value of a:", a)
```

- 5. **Bitwise Operators**
  - AND (&): Performs bitwise AND.
  - OR (|): Performs bitwise OR.
  - XOR (^): Performs bitwise XOR.
  - NOT (~): Performs bitwise NOT.
  - Left Shift (<<): Shifts bits to the left.
  - Right Shift (>>): Shifts bits to the right.

```python


# learn bit manipulation

# COMMAND ----------

a = 10  # 1010 in binary
b = 4   # 0100 in binary

bitwise_and = a & b  # 0000 in binary
bitwise_or = a | b   # 1110 in binary
bitwise_xor = a ^ b  # 1110 in binary
bitwise_not = ~a     # -11 in binary (two's complement)
left_shift = a << 2  # 101000 in binary
right_shift = a >> 2 # 0010 in binary
print("Bitwise AND:", bitwise_and)
print("Bitwise OR:", bitwise_or)
print("Bitwise XOR:", bitwise_xor)
print("Bitwise NOT:", bitwise_not)
print("Left Shift:", left_shift)
print("Right Shift:", right_shift)
```

- 6. Membership Operators
  - in: Checks if a value is present in a sequence.
  - not in: Checks if a value is not present in a sequence.
  - return True or False

```python
fruits = ["apple", "banana", "cherry"]

is_in = "banana" in fruits
is_not_in = "grape" not in fruits
print("Is 'banana' in fruits?", is_in)
print("Is 'grape' not in fruits?", is_not_in)
```
