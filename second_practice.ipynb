```python
my_string = "Hello, World!"

# 1. Slicing substrings
print("Original string:", my_string)
print("Substring from index 0 to 4 (exclusive):", my_string[0:5])  # 'Hello'
print("Substring from index 7 to end:", my_string[7:])        # 'World!'
print("Substring from beginning to index 5 (exclusive):", my_string[:5]) # 'Hello'
print("Last character:", my_string[-1])                     # '!'
print("Substring from index -6 to -1 (exclusive):", my_string[-6:-1])  # 'World'

print("\n---")

# 2. Checking for substring presence
print("Does 'World' exist in the string?", "World" in my_string)  # True
print("Does 'Python' exist in the string?", "Python" in my_string) # False

print("\n---")

# 3. Finding the index of a substring
print("Index of 'World':", my_string.find("World"))  # 7
print("Index of 'o' (first occurrence):", my_string.find("o"))    # 4
print("Index of 'Python' (not found):", my_string.find("Python")) # -1 (if not found)

# rfind() finds the last occurrence
print("Last index of 'o':", my_string.rfind("o"))   # 8

print("\n---")

# 4. Counting occurrences of a substring
print("Count of 'o':", my_string.count("o")) # 2

print("\n---")

# 5. Replacing substrings
new_string_replace = my_string.replace("World", "Python")
print("String after replacing 'World' with 'Python':", new_string_replace)

new_string_replace_all_o = my_string.replace("o", "@")
print("String after replacing all 'o' with '@':", new_string_replace_all_o)

print("\n---")

# 6. Splitting a string into a list of substrings
parts = my_string.split(", ")
print("Split by ', ':", parts) # ['Hello', 'World!']

text = "one-two-three"
parts_dash = text.split("-")
print("Split by '-':", parts_dash) # ['one', 'two', 'three']

print("\n---")

# 7. Checking for prefix and suffix
print("Starts with 'Hello':", my_string.startswith("Hello")) # True
print("Ends with 'World!':", my_string.endswith("World!"))   # True
print("Ends with 'world!':", my_string.endswith("world!"))   # False (case-sensitive)
```
